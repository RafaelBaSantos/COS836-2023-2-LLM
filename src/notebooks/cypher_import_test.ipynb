```python
from neo4j import GraphDatabase

# Connect to the Neo4j database
uri = "bolt://localhost:7687"
username = "neo4j"
password = "12345678"
driver = GraphDatabase.driver(uri, auth=(username, password))
```

```python
def wipe_database(tx):
    tx.run("MATCH (n) DETACH DELETE n")

with driver.session() as session:
    session.write_transaction(wipe_database)
```

```python
script = """CREATE (Arkanthos:Lugar {nome: 'Arkanthos', descrição: 'Um lugar místico, onde a magia flui através da terra e do ar'})
CREATE (AliançaDosReinos:Organização {nome: 'Aliança dos Reinos', descrição: 'Governo central de Arkanthos que busca manter a paz entre as diversas raças e culturas'})
CREATE (LordValerius:Personagem {nome: 'Lord Valerius', descrição: 'Destemido líder dos nobres cavalheiros da Aliança dos Reinos'})
CREATE (GuardiõesArcanos:Organização {nome: 'Guardiões Arcanos', descrição: 'Aqueles que nasceram com o dom da magia e são reverenciados como mentores poderosos'})
CREATE (Sylphina:Personagem {nome: 'Alta Arcanista Sylphina', descrição: 'Enigmática líder das Guildas de Magia'})
CREATE (TribosDasMontanhas:Organização {nome: 'Tribos das Montanhas', descrição: 'Grupo liderado pelo cacique Alderan que vive em harmonia com a natureza'})
CREATE (Carvernia:Lugar {nome: 'Carvernia', descrição: 'Cidades comerciais onde os mercadores habilidosos da Guilda dos Comerciantes negociam valiosas mercadorias'})
CREATE (Dragão:Personagem {nome: 'Dragão', descrição: 'Antigo dragão que despertou de seu sono milenar e causa destruição nas terras do sul'})
CREATE (ReinosElevados:Lugar {nome: 'Reinos Elevados', descrição: 'Terras onde o governo tem menos influência e os moradores vivem uma vida de luta e pobreza'})
CREATE (ElfosDaFloresta:Organização {nome: 'Elfos da Floresta', descrição: 'Raça com rivalidade histórica com os Anões das Montanhas'})
CREATE (AnõesDasMontanhas:Organização {nome: 'Anões das Montanhas', descrição: 'Raça com rivalidade histórica com os Elfos da Floresta'})
CREATE (Centauros:Organização {nome: 'Centauros', descrição: 'Guardiões dos bosques sagrados que defendem a harmonia entre as raças'})
CREATE (ReiDrakar:Personagem {nome: 'Rei Drakar', descrição: 'Governante dos Dragões'})
CREATE (OrdemDosGuardiões:Organização {nome: 'Ordem dos Guardiões', descrição: 'Organização de paladinos dedicados a proteger os inocentes'})
CREATE (Magnus:Personagem {nome: 'Paladino Celestial Magnus', descrição: 'Poderoso membro da Ordem dos Guardiões'})
CREATE (AprendizDeMago:Personagem {nome: 'Aprendiz de Mago', descrição: 'Jovem em busca de conhecimento e poder'})
CREATE (Cavaleiro:Personagem {nome: 'Cavaleiro', descrição: 'Destemido herói em busca de vingança pela morte de sua família'})

CREATE (Arkanthos)-[:TEM]->(AliançaDosReinos)
CREATE (AliançaDosReinos)-[:LIDERA]->(LordValerius)
CREATE (Arkanthos)-[:TEM]->(GuardiõesArcanos)
CREATE (GuardiõesArcanos)-[:LIDERA]->(Sylphina)
CREATE (Arkanthos)-[:TEM]->(TribosDasMontanhas)
CREATE (TribosDasMontanhas)-[:LIDERA]->(Alderan)
CREATE (Arkanthos)-[:TEM]->(Carvernia)
CREATE (Carvernia)-[:TEM]->(GuildaDosComerciantes)
CREATE (Arkanthos)-[:TEM]->(Dragão)
CREATE (Arkanthos)-[:TEM]->(ReinosElevados)
CREATE (Arkanthos)-[:TEM]->(ElfosDaFloresta)
CREATE (Arkanthos)-[:TEM]->(AnõesDasMontanhas)
CREATE (Arkanthos)-[:TEM]->(Centauros)
CREATE (Arkanthos)-[:TEM]->(ReiDrakar)
CREATE (Arkanthos)-[:TEM]->(OrdemDosGuardiões)
CREATE (OrdemDosGuardiões)-[:LIDERA]->(Magnus)
CREATE (Arkanthos)-[:TEM]->(AprendizDeMago)
CREATE (Arkanthos)-[:TEM]->(Cavaleiro)
"""

with driver.session() as session:
    session.run(script)
```

```python
def get_triplets(tx):
    result = tx.run("MATCH (n)-[r]->(m) RETURN n, r, m")
    return [(record["n"]["nome"], record["r"].type, record["m"]["nome"]) for record in result]

with driver.session() as session:
    triplets = session.read_transaction(get_triplets)

for triplet in triplets:
    print(triplet)
```

```python
def get_entities_and_properties(tx):
    result = tx.run("MATCH (n) RETURN n")
    return [(record["n"]["nome"], dict(record["n"].items())) for record in result]

with driver.session() as session:
    entities = session.read_transaction(get_entities_and_properties)

for entity in entities:
    print(entity)
```

```python
# Close the driver
driver.close()
```
